```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import FreqDist

nltk.data.path.append("/kaggle/input/nltk-data/nltk_data")

paragraph = """Technology is constantly evolving and reshaping the way we live. 
From smartphones to artificial intelligence, innovations are making our lives easier and more connected. 
I find it fascinating how machines can learn from data and make decisions. 
The rise of smart devices and automation is transforming industries. 
Every year, new breakthroughs bring us closer to a more digital world. 
For someone interested in problem-solving and creativity, technology offers endless possibilities."""

lower_text = paragraph.lower()
no_punct_text = ''.join(char for char in lower_text if char not in string.punctuation)

sent_tokens = sent_tokenize(no_punct_text)
word_tokens = word_tokenize(no_punct_text)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens if word not in stop_words]

print("Sentences:")
for i, sent in enumerate(sent_tokens, 1):
    print(f"{i}. {sent}")

print("\nFiltered Words (no stopwords):")
print(filtered_words)

freq_dist = FreqDist(filtered_words)

print("\nWord Frequency Distribution:")
for word, freq in freq_dist.items():
    print(f"{word}: {freq}")

import matplotlib.pyplot as plt
freq_dist.plot(20, title='Top 20 Words Frequency (Excluding Stopwords)')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

text = "The striped bats are hanging on their feet for best"

tokens = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]

print("Original Tokens (after stopword removal):")
print(filtered_tokens)

porter = PorterStemmer()
lancaster = LancasterStemmer()

porter_stems = [porter.stem(word) for word in filtered_tokens]
lancaster_stems = [lancaster.stem(word) for word in filtered_tokens]

print("\nPorter Stemmer Results:")
print(porter_stems)

print("\nLancaster Stemmer Results:")
print(lancaster_stems)

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatization Results:")
print(lemmatized)

print("\nComparison Table:")
print("{:<15} {:<15} {:<15} {:<15}".format('Original', 'Porter', 'Lancaster', 'Lemma'))
for i in range(len(filtered_tokens)):
    print("{:<15} {:<15} {:<15} {:<15}".format(filtered_tokens[i], porter_stems[i], lancaster_stems[i], lemmatized[i]))
```

```python
import re

text = """Technology is constantly evolving and reshaping the way we live. 
From smartphones to artificial intelligence, innovations are making our lives easier and more connected. 
I find it fascinating how machines can learn from data and make decisions. 
The rise of smart devices and automation is transforming industries. 
Every year, new breakthroughs bring us closer to a more digital world. 
For someone interested in problem-solving and creativity, technology offers endless possibilities."""

more_than_5 = re.findall(r'\b\w{6,}\b', text)
print("Words with more than 5 letters:\n", more_than_5)

numbers = re.findall(r'\b\d+\b', text)
print("\nNumbers:\n", numbers)

capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', text)
print("\nCapitalized Words:\n", capitalized_words)

only_alpha_words = re.findall(r'\b[a-zA-Z]+\b', text)
print("\nAlphabet-only Words:\n", only_alpha_words)

vowel_words = [word for word in only_alpha_words if re.match(r'^[aeiouAEIOU]', word)]
print("\nWords Starting with a Vowel:\n", vowel_words)
```

```python
import re

text = """Contact us at support@example.com or visit https://example.com.
Our number is 123-456-7890 or +91 9876543210. 
This state-of-the-art system isn't just fast – it's 3.14 times faster than others!"""

def clean_text_regex(text):
    text = re.sub(r'\b[\w.-]+?@\w+?\.\w+?\b', '<EMAIL>', text)
    text = re.sub(r'https?://\S+|www\.\S+', '<URL>', text)
    text = re.sub(r'\+?\d{1,3}[\s-]?\d{10}|\d{3}-\d{3}-\d{4}', '<PHONE>', text)
    return text

cleaned_text = clean_text_regex(text)

def custom_tokenizer(text):
    pattern = r"\b(?:[a-zA-Z]+(?:'[a-z]+)?(?:-[a-z]+)*|\d+\.\d+|\d+)\b"
    tokens = re.findall(pattern, text)
    return tokens

tokens = custom_tokenizer(cleaned_text)

print("Cleaned Text:\n", cleaned_text)
print("\nCustom Tokens:\n", tokens)
```
