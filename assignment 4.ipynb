```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
arr += 2
arr *= 3
arr /= 2
print(arr)
```

```python
import numpy as np
from scipy import stats
x = np.array([1,2,3,4,5,1,2,1,1,1])
most_frequent_value_x, _ = stats.mode(x)
indices_x = np.where(x == most_frequent_value_x)[0]
print(f"Most frequent value in x: {most_frequent_value_x[0]}")
print(f"Indices of most frequent value in x: {indices_x}")
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_frequent_value_y, _ = stats.mode(y)
indices_y = np.where(y == most_frequent_value_y)[0]
print(f"Most frequent value in y: {most_frequent_value_y[0]}")
print(f"Indices of most frequent value in y: {indices_y}")
```

```python
import numpy as np
your_name = np.linspace(10, 100, 25)
print(your_name)
print(f"Dimensions: {your_name.ndim}")
print(f"Shape: {your_name.shape}")
print(f"Total elements: {your_name.size}")
print(f"Data type: {your_name.dtype}")
print(f"Total bytes consumed: {your_name.nbytes}")
transposed_array = your_name.reshape(-1, 1)
print(f"Transposed array using reshape():\n{transposed_array}")
transposed_array_T = your_name.T
print(f"Transposed array using T attribute:\n{transposed_array_T}")
```
