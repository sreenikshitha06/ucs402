```python

data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

df_employees = pd.DataFrame(data)

df_employees.to_csv('employees.csv', index=False)
```

```python

print(f"Shape of the DataFrame: {df_employees.shape}")

print("\nSummary of DataFrame:")
print(df_employees.info())

print("\nDescriptive Statistics:")
print(df_employees.describe())

print("\nFirst 5 rows:")
print(df_employees.head())

print("\nLast 3 rows:")
print(df_employees.tail(3))

avg_salary = df_employees['Salary'].mean()
print(f"\nAverage salary of employees: {avg_salary}")

total_bonus = df_employees['Bonus'].sum()
print(f"Total bonus paid to all employees: {total_bonus}")

youngest_age = df_employees['Age'].min()
print(f"Youngest employee's age: {youngest_age}")

highest_rating = df_employees['Rating'].max()
print(f"Highest performance rating: {highest_rating}")

df_sorted = df_employees.sort_values(by='Salary', ascending=False)
print("\nSorted by Salary (Descending):")
print(df_sorted)

def performance_category(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df_employees['Performance'] = df_employees['Rating'].apply(performance_category)
print("\nEmployees with Performance Category:")
print(df_employees)

print("\nMissing Values in DataFrame:")
print(df_employees.isnull().sum())
df_employees.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\nRenamed Column:")
print(df_employees)

employees_filtered = df_employees[(df_employees['Years_of_Experience'] > 5) & (df_employees['Department'] == 'IT')]
print("\nEmployees with more than 5 years of experience in IT department:")
print(employees_filtered)
```

```python

df_employees['Tax'] = df_employees['Salary'] * 0.10
print("\nEmployees Dataset with Tax:")
print(df_employees)

df_employees.to_csv('modified_employees.csv', index=False)
```
