```python
import numpy as np
import matplotlib.pyplot as plt

# Q1
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(np.sum(gfg))
print(np.sum(gfg, axis=1))
print(np.sum(gfg, axis=0))

# Q2 (a)
array = np.array([10, 52, 62, 16, 16, 54, 453])
print(np.sort(array))
print(np.argsort(array))
print(np.sort(array)[:4])
print(np.sort(array)[-5:])

# Q2 (b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(array[array == array.astype(int)])
print(array[array != array.astype(int)])

# Q3
def ascii_sum(name):
    return sum(ord(char) for char in name)

X = ascii_sum('YourInitials')
sales = np.array([X, X+50, X+100, X+150, X+200])
tax_rate = ((X % 5) + 5) / 100
sales_taxed = sales * (1 - tax_rate)
sales_discounted = np.where(sales < X+100, sales * 0.95, sales * 0.90)
sales_weeks = np.vstack([sales] * 3) * np.array([1.00, 1.02, 1.04]).reshape(3, 1)
print(sales_weeks)

# Q4
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 6))
plt.plot(x, x**2, label='Y = x^2')
plt.plot(x, np.sin(x), label='Y = sin(x)')
plt.plot(x, np.exp(x), label='Y = e^x')
plt.plot(x, np.log(np.abs(x) + 1), label='Y = log(|x| + 1)')
plt.title('Mathematical Functions')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid()
plt.show()
```
