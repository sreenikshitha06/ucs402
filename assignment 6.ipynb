```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Q1
M = float(input())
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure()
plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'b-')
plt.grid()
plt.show()

# Q2
data = {'Subjects': ['Math', 'Science', 'English', 'History', 'Art'], 'Scores': [85, 90, 78, 88, 92]}
df = pd.DataFrame(data)
plt.figure()
ax = sns.barplot(x='Subjects', y='Scores', data=df, palette='coolwarm')
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', xytext=(0, 9), textcoords='offset points')
plt.grid()
plt.show()

# Q3
np.random.seed(12345)
random_data = np.random.randn(50)
cumsum_data = np.cumsum(random_data)
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(cumsum_data, 'g-')
axes[0, 1].scatter(range(50), random_data, c='b', marker='o')
plt.tight_layout()
plt.show()

# Q4
url = "https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv"
df_sales = pd.read_csv(url)
plt.figure()
plt.plot(df_sales['month_number'], df_sales['total_profit'], 'r-', marker='o')
plt.grid()
plt.show()

plt.figure()
for col in df_sales.columns[1:-1]:
    plt.plot(df_sales['month_number'], df_sales[col], marker='o')
plt.grid()
plt.show()

plt.figure()
df_sales.iloc[:, 1:].sum().plot(kind='bar', color=sns.color_palette("coolwarm", len(df_sales.columns[1:])))
plt.grid()
plt.show()
```
